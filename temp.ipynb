```python
import heapq

# 模拟的大语言模型评分函数
def model_score(string):
    # 这里用一个简单的模拟评分方法
    # 实际应用中，这里会调用真实的大模型获取分数
    return len(string)  # 假设字符串越长分数越高

# 测试数据
strings = ["apple", "banana", "orange", "pear", "grape", "kiwi", "melon"]
k = 3  # 保留top3的KV Cache

# 存储 (-score, index) 以便最小堆实现最大堆
top_k_heap = []

print("所有字符串及其分数:")
for index, string in enumerate(strings):
    score = model_score(string)
    print(f"索引 {index}: '{string}' -> 分数 {score}")
    
    # 使用元组存储分数和索引
    item = (-score, index)  # 负分用于最小堆模拟最大堆
    
    if len(top_k_heap) < k:
        heapq.heappush(top_k_heap, item)
    else:
        heapq.heappushpop(top_k_heap, item)

# 获取topk的索引
topk_indices = [item[1] for item in top_k_heap]
topk_scores = [-item[0] for item in top_k_heap]  # 恢复原始分数

print("\nTop-K 结果:")
print(f"保留的KV Cache索引: {topk_indices}")
print(f"对应的字符串: {[strings[i] for i in topk_indices]}")
print(f"对应的分数: {topk_scores}")

# 验证是否正确
expected_top = sorted([(model_score(s), i) for i, s in enumerate(strings)], reverse=True)[:k]
expected_indices = [i for (score, i) in expected_top]
print("\n验证:")
print(f"预期的Top-K索引: {expected_indices}")
print("测试通过!" if set(topk_indices) == set(expected_indices) else "测试失败!")
```

```python
from tqdm import tqdm
fnames = ['chunk_4k_1.jsonl', 'chunk_4k_4.jsonl', 'chunk_8k_1.jsonl', 'chunk_8k_4.jsonl']
task_name = 'kv_retrieval'
# fnames = ['chunk_4k_4.jsonl']
for fname in fnames:
    _, window, num = fname.split('_')
    num = num.split('.')[0]
    tokenizer = tokenizer1 if window == '4k' else tokenizer2
    print('='*20)
    print(fname)
    pth = f'index_data/InfiniteBench/{task_name}/{fname}'
    data = load_jsonl(pth)
    # find gt
    for item in tqdm(data):
        context_length = [len(tokenizer.encode(context, add_special_tokens=False)) for context in item['context_list']]
        item['context_length'] = context_length
        gt_idx = -1
        ans = ": \""+item['answer'][0]+"\""
        for i, context in enumerate(item['context_list']):
            if ans in context:
            # if ' '+ans in context or '['+ans in context:
                gt_idx = i
                break
        assert gt_idx != -1, f'gt not found in context: {ans} {context}'
        item['gt_idx'] = gt_idx
    # stat
    nums = [len(item['context_list']) for item in data]
    avg_nums = sum(nums) / len(nums)
    length = [item['context_length'] for item in data]
    length = [l for lens in length for l in lens]
    avg_length = sum(length) / len(length)
    print('='*20)
    print('length.avg: ', avg_length)
    print('nums.avg: ', avg_nums)
    print('='*20)
    save_jsonl(f'retrieval_data/{task_name}/chunk_num{num}_{window}.jsonl', data)
```

```python
REGEX_EN = r'([\n.!?;]\s*)'
import re

def chunk_by_task(tokenizer, text, length):
    sentences = re.split(REGEX_EN, text)
    chunks = split_long_context(tokenizer, sentences, length)
    return chunks

def split_long_context(tokenizer, sentences, token_nums):
    chunks = []
    sentence_length = [len(tokenizer.encode(s, add_special_tokens=False)) for s in sentences]
    
    chunk, cum_len = '', 0
    for i, length in enumerate(sentence_length):
        if cum_len + length <= token_nums:
            chunk += sentences[i]
            cum_len += length
        else:
            if chunk:  # 避免添加空块
                chunks.append(chunk)
            chunk, cum_len = sentences[i], length  # 开始新的块，包含当前句子
    if chunk:  # 添加最后一个块
        chunks.append(chunk)
    return chunks
```
